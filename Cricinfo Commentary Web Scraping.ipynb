```python
def obtain_match_links_by_year(year):
    records = requests.get(lst_of_years[year - 2005])
    souped_record = BeautifulSoup(records.content, 'html.parser')
    lst_of_all_links = souped_record.find_all("a", class_="data-link")
    matches_of_given_year = []
    for link in lst_of_all_links:
        if link.get("href").find("/engine/match") != -1:
            matches_of_given_year += ["http://stats.cricinfo.com" + link.get("href")]
    return matches_of_given_year
```

```python
lst_of_all_matches = []
for num in range(2005, 2019):
    lst_of_all_matches += (obtain_match_links_by_year(num))
```

```python
def match_link_to_cm(lst):
    cm_links = []
    counter = 1
    for link in lst:
        match_link = requests.get(link)
        souped_match = BeautifulSoup(match_link.content, 'html.parser')
        cm = souped_match.find("a", class_="commentary react-router-link")
        if cm is not None:
            cm_links += ["http://cricinfo.com" + cm.get("href")]
        if counter % 100 == 0:
            print(counter)
        counter += 1
    return cm_links
```

```python
def clean(match):
    first_data = match[0][0]
    first_cm = match[0][1]
    second_data = match[1][0]
    second_cm = match[1][1]
    
    first_data_to_ret = []
    first_cm_to_ret = []
    second_data_to_ret = []
    second_cm_to_ret = []

    first_index = 0
    second_index = 0
    
    for item in first_cm:
        temp = item.strip()
        last_char = temp[len(temp)-1]
        if last_char != ',':
            first_cm_to_ret.append(item)
            first_data_to_ret.append(first_data[first_index])
            first_index += 1
            
    for item in second_cm:
        temp = item.strip()
        last_char = temp[len(temp)-1]
        if last_char != ',':
            second_cm_to_ret.append(item)
            second_data_to_ret.append(second_data[second_index])
            second_index += 1
            
    if len(first_data_to_ret) == 0:
        return None
            
    return [[first_data_to_ret, first_cm_to_ret], [second_data_to_ret, second_cm_to_ret]]
```

```python
cleaned_lst_of_data_and_cm = [clean(phrases) for phrases in lst_of_data_and_cm if clean(phrases) is not None]
```

```python
def classify(phrase):
    words = ["defend", "leave", "hook", "pull", "drive", "sweep", "cut", "slog", "punch", "push", "work", "glance"]
    for word in words:
        if word in phrase:
            return 1
    return 0
```

```python
def coverage():
    classified_matches = 0
    total_matches = 0
    phrases_not_classified = []
    phrases_classified = []
    for match in cleaned_lst_of_data_and_cm:
        first_cm = match[0][1]
        second_cm = match[1][1]
        for cm_phrase in first_cm:
            classify_phrase = classify(cm_phrase)
            classified_matches += classify_phrase
            if classify_phrase == 0:
                phrases_not_classified.append(cm_phrase)
            else:
                phrases_classified.append(cm_phrase)
            total_matches += 1
        for cm_phrase in second_cm:
            classify_phrase = classify(cm_phrase)
            classified_matches += classify_phrase
            if classify_phrase == 0:
                phrases_not_classified.append(cm_phrase)
            else:
                phrases_classified.append(cm_phrase)
            total_matches += 1
            
    return phrases_classified

phrases_to_classify = coverage()
```

```python
#runs, balls, wickets, boundaries
shot_dictionary = {"defend" : [0, 0, 0, 0],  "leave" : [0, 0, 0, 0], "hook" : [0, 0, 0, 0], "pull" : [0, 0, 0, 0], "drive" : [0, 0, 0, 0], 
                   "sweep" : [0, 0, 0, 0], "cut" : [0, 0, 0, 0], "slog" : [0, 0, 0, 0], "punch" : [0, 0, 0, 0], "push" : [0, 0, 0, 0],
                   "work" : [0, 0, 0, 0], "glance" : [0, 0, 0, 0], "flick" : [0, 0, 0, 0], "squeeze" : [0, 0, 0, 0], "steer" : [0, 0, 0, 0], 
                  "loft" : [0, 0, 0, 0], "tap" : [0, 0, 0, 0], "uppercut" : [0, 0, 0, 0], "nudge" : [0, 0, 0, 0]}

def reset_shot_dict():
    shot_dictionary = {"defend" : [0, 0, 0, 0],  "leave" : [0, 0, 0, 0], "hook" : [0, 0, 0, 0], "pull" : [0, 0, 0, 0], "drive" : [0, 0, 0, 0], 
                   "sweep" : [0, 0, 0, 0], "cut" : [0, 0, 0, 0], "slog" : [0, 0, 0, 0], "punch" : [0, 0, 0, 0], "push" : [0, 0, 0, 0],
                   "work" : [0, 0, 0, 0], "glance" : [0, 0, 0, 0], "flick" : [0, 0, 0, 0], "squeeze" : [0, 0, 0, 0], "steer" : [0, 0, 0, 0], 
                  "loft" : [0, 0, 0, 0], "tap" : [0, 0, 0, 0], "uppercut" : [0, 0, 0, 0], "nudge" : [0, 0, 0, 0]}
```

```python
def classify_phrase(phrase):
    
    ordered_list = ["hook", "pull", "slog", "loft", "sweep", "drive", 
                    "defend", "uppercut", "cut", "punch", "push", "work", "glance", 
                   "flick", "squeeze", "steer", "tap", "nudge"] #excluding "leave"
    list_of_scores = ["no run", "1 run", "2 runs", "3 runs", "FOUR runs", "6 runs", "SIX", "OUT"]
    score_dict = {"no run" : 0, "1 run" : 1, "2 runs" : 2, "3 runs" : 3, 
                  "FOUR runs" : 4, "SIX" : 6, "6 runs" : 6, "OUT" : 0, 
                 "1 leg bye" : 1, "2 leg byes" : 2, "3 leg byes" : 3, "4 leg byes" : 4, 
                 "1 bye" : 1, "2 byes" : 2, "3 byes" : 3, "4 byes" : 4}
    
    reset_shot_dict()
    
    for shot_type in ordered_list:
        if shot_type in phrase:
            if shot_type == "sweep" and "sweeper" in phrase:
                continue
            if shot_type == "cut" and ("cutter" in phrase or "cuts it off" in phrase or "cut it off" in phrase):
                continue
            score = ""
            for scores in list_of_scores:
                if scores in phrase:
                    score = scores
                    break
                
            num_score = 0
            if len(score) > 1:
                shot_dictionary[shot_type][1] += 1
                if score == "OUT":
                    shot_dictionary[shot_type][2] += 1
                else:
                    num_score = score_dict[score]
                    shot_dictionary[shot_type][0] += num_score
                    if num_score >= 4:
                        shot_dictionary[shot_type][3] += 1
            return (shot_type, num_score, (num_score >= 4))
```

```python
from operator import add
#list(map(add, list1, list2) )

category_1 = ["defend", "leave", "nudge", "tap"]
category_2 = ["cut", "drive", "flick", "glance", "pull", "punch", "push", 
              "squeeze", "steer", "work", "sweep"]
category_3 = ["hook", "loft", "slog", "uppercut"]

category_shot_dictionary = {"one" : [0, 0, 0], "two" : [0, 0, 0], "three" : [0, 0, 0]}

category_shot_dictionary["one"][0] = sum([shot_dictionary[shot][0] for shot in category_1])
category_shot_dictionary["one"][1] = sum([shot_dictionary[shot][1] for shot in category_1])
category_shot_dictionary["one"][2] = sum([shot_dictionary[shot][2] for shot in category_1])

category_shot_dictionary["two"][0] = sum([shot_dictionary[shot][0] for shot in category_2])
category_shot_dictionary["two"][1] = sum([shot_dictionary[shot][1] for shot in category_2])
category_shot_dictionary["two"][2] = sum([shot_dictionary[shot][2] for shot in category_2])

category_shot_dictionary["three"][0] = sum([shot_dictionary[shot][0] for shot in category_3])
category_shot_dictionary["three"][1] = sum([shot_dictionary[shot][1] for shot in category_3])
category_shot_dictionary["three"][2] = sum([shot_dictionary[shot][2] for shot in category_3])
category_shot_dictionary
```

```python
#percentage non-wickets, runs per non-wicket ball, percentage wicket
percentage_shot_dictionary = {}
for key in category_shot_dictionary:
    key_lst = []
    non_wicket_balls = category_shot_dictionary[key][1] - category_shot_dictionary[key][2]
    key_lst.append(non_wicket_balls / category_shot_dictionary[key][1])
    
    key_lst.append(category_shot_dictionary[key][0] / non_wicket_balls)
    key_lst.append(category_shot_dictionary[key][2] / category_shot_dictionary[key][1])
    
    percentage_shot_dictionary[key] = key_lst
```

```python
from pulp import *
problem = LpProblem("Shot Selection", LpMaximize)

one_prob = LpVariable("one_prob", 0, 1)
two_prob = LpVariable("two_prob", 0, 1)
three_prob = LpVariable("three_prob", 0, 1)
prob_constraint = (one_prob + two_prob + three_prob == 1)
problem += prob_constraint

worst_case = LpVariable("worst_case")
problem += worst_case

for i in range(1000):
    total = (one_prob * one_row[i]) + (two_prob * two_row[i]) + (three_prob * three_row[i])
    constraint = (total >= worst_case)
    problem += constraint
    
problem.solve()
```
